## Task 4: Build an EV recommendation class. The class should allow users to input their budget, desired range, and battery capacity. The class should then return the top three EVs matching their criteria. 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

dataset=pd.read_csv("FEV-data.csv")

class ev_recommendation:
    def __init__(self, data):
        self.data = data
        
    def user_inputs(self):
        self.budget=int(input("Enter the Budget which should be Between 82050 to 794000 = "))
        self.desired_range=int(input("Enter your desired range which should be Between 148 to 652 = "))
        self.battery_capacity=int(input("Enter the battery capacity which should be between 17.6 to 100 = "))
    
    def recommendations(self):
        filtered_dataset = self.data[(self.data['Minimal price (gross) [PLN]'] <= self.budget) 
        & (self.data['Range (WLTP) [km]'] >= self.desired_range)
        & (self.data['Battery capacity [kWh]'] >= self.battery_capacity)
    ]
        sort_price=(filtered_dataset.sort_values(by='Minimal price (gross) [PLN]',ascending=True))
        
        return sort_price

ev_recommend=ev_recommendation(dataset)
ev_recommend.user_inputs()
result=ev_recommend.recommendations()

result
```

## Task 5: Inferential Statistics – Hypothesis Testing:
## Test whether there is a significant difference in the average Engine power [KM] of vehicles manufactured by two leading manufacturers i.e. Tesla and Audi.
## What insights can you draw from the test results? Recommendations and Conclusion: 
## Provide actionable insights based on your analysis.(Conduct a two sample t-test using ttest_ind from scipy.stats module) 

```python
import pandas as pd
import numpy as np
import scipy.stats as stats

df=pd.read_csv("FEV-data.csv")
# filter the data for manufacturers Tesla and Audi
# And check the difference

tesla_make=df[df['Make'] == "Tesla"]["Engine power [KM]"]
audi_make=df[df['Make'] == "Audi"]["Engine power [KM]"]
# audi_make
print(f"Mean of Engine power of tesla is {tesla_make.mean()}")
print(f"Mean of Engine power of audi is {audi_make.mean()}")
```

```python
# H0 = Mean Engline power of Tesla = Audi
# H1 = Mean Engile power of Tesla != Audi
# Perform independent t test
t_stat, p_value = stats.ttest_ind(tesla_make,audi_make)
alpha=0.05
print(f"T-Statistic: {t_stat:.2f}")
print(f"P-Value: {p_value:.2f}")
if p_value < alpha:
    print("Reject the null hypothesis")
    print("There is an significant difference between average engine power between Tesla & Audi")
else:
    print("Fail to Reject the null hypothesis")
    print("There is no an significant difference between average engine power between Tesla & Audi")
```

1. So hear we first import all the important library we need to perform the t test like scipy.stats,pandas etc.
2. Then we imported the data.
3. Then we filter the data for manufacturers "Tesla" and "Audi"
4. So if we have Null hypothesis H0 = Mean Engline power of Tesla = Audi and Alternative hypotheses  H1 = Mean Engile power of Tesla != Audi.
5. Now we Perform independent t test using ttest_ind. and assuming the alpha=0.05(5%) which means ignoring the 5% change.
6. Even though Tesla’s average engine power (533 KM) is higher than Audi’s (392 KM), the result of the two-sample t-test (p = 0.12) shows that this difference is not statistically significant at the 0.05 level, likely due to small sample size and high variance in Tesla's data.
